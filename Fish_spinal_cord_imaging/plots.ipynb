# zstd

```python
runID_list = ['double_fish_1_before','double_fish_1_after',
              'double_fish_2_before','double_fish_2_after',
              'double_fish_3_before','double_fish_3_after',
              'single_cutA_before','single_cutA_after',
              'single_cutM_before','single_cutM_after',
              'single_cutP_before','single_cutP_after']
all_data=[np.load(runID+"_compression_rate_zstd.npy") for runID in runID_list]

import matplotlib.pyplot as plt
import numpy as np

# 创建一些示例数据
categories = ['fish_1', 'fish_2', 'fish_3','fish_4','fish_5','fish_6']
group1_values = all_data[0::2]
group2_values = all_data[1::2]

# 设置柱状图的宽度
bar_width = 0.35

# 计算每组柱状图的 x 坐标
bar_positions_group1 = np.arange(len(categories))
bar_positions_group2 = bar_positions_group1 + bar_width

# 画柱状图
plt.bar(bar_positions_group1, group1_values, width=bar_width, label='pre-ablation')
plt.bar(bar_positions_group2, group2_values, width=bar_width, label='post-ablation')

# 设置 x 轴标签和标题
plt.xlabel('subjects')
plt.ylabel('compression ratio')
plt.title('zstd compressor')

# 设置 x 轴刻度标签
plt.xticks(bar_positions_group1 + bar_width / 2, categories)

# 添加图例
plt.legend(loc='lower right')

# 显示图形
plt.show()
```
